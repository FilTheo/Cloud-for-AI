```python
import pandas as pd
```

```python
path = '/home/filtheo/Cloud-for-AI/workshop_ml_2025/data/raw/sales_train_validation.csv'
path_2 = '/home/filtheo/Cloud-for-AI/workshop_ml_2025/data/raw/calendar.csv'
main_df = pd.read_csv(path)
calendar_df = pd.read_csv(path_2)
```

```python
# keep only relevant cols for now
calendar_df = calendar_df[['date', 'd']].drop_duplicates()
```

```python

conveted_df = main_df.copy()

# focus on foods
conveted_df = conveted_df[conveted_df['cat_id'] == 'FOODS']

# focus on stores
to_drop = ['id', 'item_id', 'dept_id', 'cat_id','state_id']
conveted_df = conveted_df.drop(to_drop, axis = 1)

# aggregate by store
conveted_df = conveted_df.groupby(['store_id']).sum().reset_index()
conveted_df
```

```python
# Create mapping from d to date
date_mapping = dict(zip(calendar_df['d'], calendar_df['date']))

# Rename columns in conveted_df
conveted_df = conveted_df.rename(columns=date_mapping)
conveted_df
```

```python
conveted_df = conveted_df.melt(id_vars=['store_id'], var_name='date', value_name='y').rename(columns={'store_id': 'unique_id'})
```

```python
# keep 2.5 years of data
conveted_df = conveted_df[conveted_df['date'] >= '2013-01-01']
```

```python
save_path = '/home/filtheo/Cloud-for-AI/workshop_ml_2025/data/converted_df.csv'
conveted_df.to_csv(save_path, index=False)
```

```python
conveted_df_2 = main_df.copy()

# focus on foods
conveted_df_2 = conveted_df_2[(conveted_df_2['cat_id'] == 'FOODS')]

top_50_food_items = conveted_df_2.groupby(['item_id']).sum(numeric_only=True).sum(axis = 1).sort_values(ascending=False).head(50).index

conveted_df_2 = conveted_df_2[conveted_df_2['item_id'].isin(top_50_food_items)]

# focus on these items only
to_drop = ['id','dept_id', 'cat_id','state_id', 'store_id']
conveted_df_2 = conveted_df_2.drop(to_drop, axis = 1)

# aggregate by store
conveted_df_2 = conveted_df_2.groupby(['item_id']).sum().reset_index()
conveted_df_2
```

```python
# Create mapping from d to date
date_mapping = dict(zip(calendar_df['d'], calendar_df['date']))

# Rename columns in conveted_df
conveted_df_2 = conveted_df_2.rename(columns=date_mapping)
conveted_df_2
```

```python
conveted_df_2 = conveted_df_2.melt(id_vars=['item_id'], var_name='date', value_name='y').rename(columns={'item_id': 'unique_id'})
```

```python
# keep 2.5 years of data
conveted_df_2 = conveted_df_2[conveted_df_2['date'] >= '2013-01-01']
```

```python
save_path = '/home/filtheo/Cloud-for-AI/workshop_ml_2025/data/converted_df_2.csv'
conveted_df_2.to_csv(save_path, index=False)
```
